## Функции

```python
df[['title', 'release_year', 'rating']]
```

```python
df_cols = df.loc[:, ['title', 'release_year', 'rating']]
```

```python
# dic = {
#     'title': ['release_year', 'rating']
# }

def dic_names(df_cols):
    dic = {}

    for index in range(len(df_cols)):
        promez_df = df_cols.iloc[index]

        title = promez_df['title']
        release_year = promez_df['release_year']
        rating = promez_df['rating']

        dic[title] = [release_year, rating]
        
    return dic
```
